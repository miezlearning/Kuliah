> # Tampilkan jumlah label dari semua attribute yang bernilai object

```python
# Konversi Atribut Penghasilan menjadi object
df['Penghasilan'] = df['Penghasilan'].astype(object)

objek = df.select_dtypes(include=['object'])


if not objek.empty:
    for column in objek:
        print(f"Attribute: {column}")
        print(df[column].value_counts(), "\n")
else:
    print(f"Tidak ada atribut dengan tipe object.")


df['Penghasilan'] = df['Penghasilan'].astype(int)
```

> # Visualisasikan dataset dengan salah satu grafik. Pilih grafik yang paling cocok untuk memvisualisasikan data tersebut.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Penghasilan', data=df)
plt.title('Distribusi Penghasilan')
plt.xlabel('Penghasilan')
plt.ylabel('Jumlah Pembeli')
plt.show()
```

> # Tampilkan korelasi dataset menggunakan heatmap.

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antar Variabel')
plt.show()
```
